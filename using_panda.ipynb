# 3. Update

### 3.1 Dropping Rows

```python
# to display columns that has null values

df.isnull().sum() # to display the number of null values in the columns
```

```python
# to drop the columns with null values
df.dropna(inplace=True) # axis=1 is used to drop the columns
```

```python
df.isnull().sum() # to display the number of null values in the columns
```

### 3.2 Dropping Columns

```python
df.drop('Area code', axis = 1) # to drop the column. This dropped the 'Area code' column
```

```python
df['New column'] = df['Total eve charge'] + df['Total night charge'] # to add a new column to the DataFrame
```

```python
df.head() # to display the first 5 rows of the DataFrame with the 'New column'
```

### 3.3 Creating Calculated Columns

```python
df['New column'] = 100 # this will assign 100 to all the rows in the 'New column'
```

```python
df.head()
```

### 3.4 Updating an Entire Column

```python
df.iloc[0, -1] = 10 # this will assign 10 to the 1st row  in the 'New column'
```

```python
df.head()
```

### 3.6 Condition based Updating using Apply

```python
# Check if 'Churn' column exists in the DataFrame
if 'Churn' in df.columns:
	df['Churn Binary'] = df['Churn'].apply(lambda x: 1 if x == True else 0) # to convert the categorical column to binary
else:
	print("Column 'Churn' does not exist in the DataFrame")
```

```python
df.head()
```
